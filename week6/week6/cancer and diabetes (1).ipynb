```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_validate
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import classification_report,confusion_matrix
```

```python
data=pd.read_csv("cancer.csv.csv")
```

```python
data.shape
```

```python
data.head(5)
```

```python
data.isnull().sum()
```

```python
data.columns
```

```python
data=data.drop('Unnamed: 32',axis=1)
```

```python
data.head(5)
```

```python
x_data=data.drop('diagnosis',axis=1)
```

```python
x_data
```

```python
lm=LabelEncoder()
```

```python
data['diagnosis']=lm.fit_transform(data['diagnosis'])
```

```python
data['diagnosis']
```

```python
y_data=data['diagnosis']
```

```python
y_data
```

```python
test_sizes = [0.3, 0.4, 0.5]
results = []

for test_size in test_sizes:
    print(f'Processing test size: {test_size}')
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(x_data, y_data, test_size=test_size, random_state=42)
    model = GaussianNB()
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Classification report and confusion matrix
    report = classification_report(y_test, y_pred, output_dict=True)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    # Append results
    results.append({
        'Test Size': test_size,
        'Classification Report': report,
        'Confusion Matrix': conf_matrix.tolist()
    })

# Convert results to DataFrame and save to Excel
results_df = pd.DataFrame(results)

# Save DataFrame to Excel
excel_path = 'Naive_Bayes_Cancer_Results1.xlsx'
results_df.to_excel(excel_path, index=False)

print(f'Results saved to {excel_path}')

    
```

```python
df=pd.read_csv('diabetes.csv')
```

```python
df.columns
```

```python
X = df.drop(columns=['Outcome'])
y = df['Outcome']
```

```python
test_sizes = [0.3, 0.4, 0.5]
results = []

for test_size in test_sizes:
    print(f'Processing test size: {test_size}')
    
    # Split data
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
    
    # Initialize and train Naive Bayes classifier
    model = GaussianNB()
    model.fit(X_train, y_train)
    
    # Predict
    y_pred = model.predict(X_test)
    
    # Classification report and confusion matrix
    report = classification_report(y_test, y_pred, output_dict=True)
    conf_matrix = confusion_matrix(y_test, y_pred)
    
    # Append results
    results.append({
        'Test Size': test_size,
        'Classification Report': report,
        'Confusion Matrix': conf_matrix.tolist()
    })

# Convert results to DataFrame and save to Excel
results_df = pd.DataFrame(results)

# Save DataFrame to Excel
excel_path = 'Naive_Bayes_Diabetes_Results.xlsx'
results_df.to_excel(excel_path, index=False)

print(f'Results saved to {excel_path}')
```
