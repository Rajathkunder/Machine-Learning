```python
#a) Simple Scatter Plot
import matplotlib.pyplot as plt
import numpy as np


x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

plt.scatter(x, y, color='blue')
plt.title('Simple Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#b) Sine curve plot
import matplotlib.pyplot as plt
import numpy as np


x = np.arange(0, 2 * np.pi, 0.1)
y = np.sin(x)

plt.plot(x, y, label='Sine Curve')
plt.title('Sine Curve Plot')
plt.xlabel('x-axis')
plt.ylabel('y = sin(x)')
plt.legend()
plt.show()
```

```python
#c) Data generation using random function
import matplotlib.pyplot as plt
import numpy as np


x = np.random.rand(10)
y = np.random.rand(10)

plt.scatter(x, y, color='green')
plt.title('Random Data Scatter Plot')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#d) Histograms, bar plots, box plot
#i.Histograms
import matplotlib.pyplot as plt
import numpy as np


data = np.random.randn(10)

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
#d) Histograms, bar plots, box plot
#ii.Bar plots
import matplotlib.pyplot as plt


categories = ['A', 'B', 'C', 'D']
values = [5, 7, 3, 9]

plt.bar(categories, values, color='orange')
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
#d) Histograms, bar plots, box plot
#iii.Box plot
import matplotlib.pyplot as plt
import numpy as np


data = np.random.rand(10, 2)

plt.boxplot(data)
plt.title('Box Plot')
plt.xlabel('Sample Index')
plt.ylabel('Value')
plt.show()
```
