```python
#once sheet is created with required data 
#populate all fields and values

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt


df = pd.read_csv('StudentDataML.csv') 

# (c) Compute the Mean, Median, Mode, and 5-number summary of all Marks Columns
mean = df[['M1', 'M2', 'M3', 'M4', 'M5']].mean()
median = df[['M1', 'M2', 'M3', 'M4', 'M5']].median()
mode = df[['M1', 'M2', 'M3', 'M4', 'M5']].mode().iloc[0]
summary = df[['M1', 'M2', 'M3', 'M4', 'M5']].describe().loc[['min', '25%', '50%', '75%', 'max']]

print("Mean:\n", mean)
print("\nMedian:\n", median)
print("\nMode:\n", mode)
print("\n5-Number Summary:\n", summary)
```

```python
# (d) Draw a box plot for all numeric columns
df[['M1', 'M2', 'M3', 'M4', 'M5']].plot(kind='box', title='Box Plot for Student Marks')
plt.show()
```

```python
# (e) Draw various possible plots for student marks in each subject
df[['M1', 'M2', 'M3', 'M4', 'M5']].plot(kind='line', marker='o', title='Line Plot for Student Marks')
plt.xlabel('Student Index')
plt.ylabel('Marks')
#plt.grid(True)
plt.show()
```

```python
df[['M1', 'M2', 'M3', 'M4', 'M5']].plot(kind='bar', title='Bar Plot for Student Marks')
plt.xlabel('Student Index')
plt.ylabel('Marks')
#plt.grid(True)
plt.show()
```

```python
df[['M1', 'M2', 'M3', 'M4', 'M5']].plot(kind='hist', alpha=0.5, bins=10, title='Histogram for Student Marks')
plt.xlabel('Marks')
plt.ylabel('Frequency')
#plt.grid(True)
plt.show()
```

```python
# (f) Display the outlier values if any in each of the students' marks columns
outliers = {}
for col in ['M1', 'M2', 'M3', 'M4', 'M5']:
    q1 = df[col].quantile(0.25)
    q3 = df[col].quantile(0.75)
    iqr = q3 - q1
    lower_bound = q1 - 1.5 * iqr
    upper_bound = q3 + 1.5 * iqr
    outliers[col] = df[(df[col] < lower_bound) | (df[col] > upper_bound)][col]

print("Outliers:\n", outliers)
```
