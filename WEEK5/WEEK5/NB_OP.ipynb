```python
#play_tennis dataset
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, mean_squared_error
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
import numpy as np

# Load the dataset
data = pd.read_csv('play_tennis.csv')

# Display first few rows of the dataset (for understanding)


# Split features (X) and target (y)
X = data[['outlook', 'temp', 'humidity', 'wind']]
y = data['play']

# Convert categorical data to numeric
X = pd.get_dummies(X)

# Encode the target variable using LabelEncoder
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=0)

# Initialize the Naive Bayes model
model = GaussianNB()

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate accuracy, MSE, and misclassification rate
accuracy = accuracy_score(y_test, y_pred)

misclassification_rate = 1 - accuracy  # Misclassification rate is 1 minus the accuracy

print(f"Accuracy: {accuracy * 100:.2f}%")

print(f"Misclassification Rate: {misclassification_rate * 100:.2f}%")

# Generate indices for plotting
index = np.arange(len(y_test))
bar_width = 0.35

# Plot bars for actual and predicted values
plt.bar(index, y_test, bar_width, color='blue', label='Actual')
plt.bar(index + bar_width, y_pred, bar_width, color='red', label='Predicted')

plt.title('Actual vs Predicted Play Tennis')
plt.xlabel('Sample index')
plt.ylabel('Play (Encoded as 0/1)')

plt.legend()

plt.tight_layout()
plt.show()
```

```python
#titanic dataset
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# Load the Titanic dataset
data = sns.load_dataset('titanic')

# Display first few rows of the dataset (for understanding)
print(data.head())

# Select relevant features and target
X = data[['pclass', 'sex', 'age', 'sibsp', 'parch', 'fare']]
y = data['survived']

# Handle missing values by filling them with median for numeric columns
X['age'].fillna(X['age'].median(), inplace=True)
X['fare'].fillna(X['fare'].median(), inplace=True)

# Convert categorical data to numeric
X = pd.get_dummies(X, columns=['sex', 'pclass'], drop_first=True)

# Encode the target variable
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Initialize the Naive Bayes model
model = GaussianNB()

# Train the model
model.fit(X_train, y_train)

# Predict on the test set
y_pred = model.predict(X_test)

# Calculate accuracy and misclassification rate
accuracy = accuracy_score(y_test, y_pred)
misclassification_rate = 1 - accuracy

print(f"Accuracy: {accuracy * 100:.2f}%")
print(f"Misclassification Rate: {misclassification_rate * 100:.2f}%")

# Generate indices for plotting
index = np.arange(len(y_test))
bar_width = 0.35


```
