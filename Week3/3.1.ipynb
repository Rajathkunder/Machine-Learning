```python
import numpy as np
import matplotlib.pyplot as plt

# Sample Data
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

#it is data fro scatter and box plot
#in below syntx , we are using normal standard distribution to generate random values, 0 is the mean and 1 is the SD, it outputs an array for 10 values.
x = np.random.normal(0, 1, 10)
y = np.random.normal(0, 1, 10)
#it is data for histogram
#belw, generate rando valus using guassian distributin , 0 is mean and 1 is SD, 10 no ele.
data = np.random.randn(1000)
```

```python
plt.hist(data, color='blue', edgecolor='black')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.bar(categories, values, color='green')
plt.title('Bar Chart')
plt.xlabel('Category')
plt.ylabel('Values')
plt.show()
```

```python
plt.pie(values, labels=categories, autopct='%1.1f%%', colors=['red', 'green', 'blue', 'orange'])
plt.title('Pie Chart')
plt.show()
```

```python
plt.scatter(x, y, color='purple', alpha=0.5)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
plt.boxplot([x, y], labels=['X', 'Y'])
plt.title('Box Plot')
plt.show()
```
